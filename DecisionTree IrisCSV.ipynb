```python
import numpy as np
import pandas as pd
from sklearn import tree
```

```python
pokemonDataset = pd.read_csv('dataset_pokemon.csv',delimiter=';',header=0)
```

```python
pokemonDataset.head()
```

```python
pokemonDataset["Species"]=pd.factorize(pokemonDataset.Species)[0]
```

```python
pokemonDataset.head()
```

```python
print(pokemonDataset)
```

```python
pokemonDataset = pokemonDataset.drop(labels="Id",axis=1)
```

```python
print(pokemonDataset)
```

```python
pokemonDataset = pokemonDataset.to_numpy()
```

```python
print(pokemonDataset)
```

```python
dataTraining = np.concatenate((pokemonDataset[0:40, :], pokemonDataset[50:90, :]),axis=0)
dataTesting = np.concatenate((pokemonDataset[40:50, :], pokemonDataset[90:100, :]),axis=0)
```

```python
print(dataTesting)
len(dataTesting)
```

```python
inputTraining = dataTraining[:,0:4]
inputTesting = dataTesting[:,0:4]
labelTraining = dataTraining[:,4]
labelTesting = dataTesting[:,4]
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model = model.fit(inputTraining,labelTraining)
```

```python
hasilPrediksi = model.predict(inputTesting)
print("Label Sebenarnya : ",labelTesting)
print("Hasil Prediksi : ",hasilPrediksi)
```

```python
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("Prediksi Benar : ",prediksiBenar,"data")
print("Prediksi Benar : ",prediksiSalah,"data")
print("Akurasi : ", prediksiBenar/(prediksiBenar+prediksiSalah)*100,"%")
```
